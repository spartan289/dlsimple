```python
import pandas as pd
data = pd.read_csv(r'C:\Users\student\Downloads\winequality-red.csv')
```

```python
data
```

```python
y = data['quality']
X = data.drop(columns=['quality'])
```

```python
X.shape
```

```python
from sklearn.preprocessing import StandardScaler
sc = StandardScaler()
X = sc.fit_transform(X)

from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X,y,test_size=0.2,random_state=4)
```

```python
import keras
from keras.models import Sequential
from keras.layers import Dense
import tensorflow as tf
```

```python
ann = Sequential()
ann.add(Dense(units=11,activation='relu'))
ann.add(Dense(units=15,activation='relu'))

ann.add(Dense(units=9,activation='softmax'))




opt = tf.keras.optimizers.Adam()
loss = tf.keras.losses.SparseCategoricalCrossentropy()
ann.compile(loss= loss
            ,optimizer=opt,
            metrics=['accuracy'])
h = ann.fit(X_train, y_train, batch_size = 50, validation_data = (X_test,y_test), epochs = 100)
```
